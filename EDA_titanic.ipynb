```python
def compute_age(cols):
    Age = cols[0]
    Corelated = cols[1]
    
    if pd.isnull(Age):
        if Corelated == 1:
            return 37
        
        if Corelated == 2:
            return 29
        
        else:
            return 24
        
    else:
        return Age
        
```

```python
df['Age']=df[['Age','Pclass']].apply(compute_age,axis=1)
```
